# 🎤 XTTS Voice Cloning on Google Colab
Clone any voice from a short sample and generate speech using Coqui XTTS v2.

⚠️ Voice sample should be WAV format, 16kHz, mono, and under 30 seconds.

```python
# ✅ Install dependencies
!pip install -q TTS
!apt-get install -y sox ffmpeg
```

```python
# ✅ Upload your voice sample (short .wav file) and define your script
from google.colab import files
uploaded = files.upload()  # Upload a short voice sample (e.g., custom_voice.wav)
```

```python
# ✅ Generate speech using XTTS (Coqui)
from TTS.api import TTS

voice_path = list(uploaded.keys())[0]  # Use uploaded file
output_file = "cloned_voice.wav"
text = "Welcome to VelociByte Hustle! This is your cloned voice speaking from Google Colab."

# Load XTTS model
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2", progress_bar=False, gpu=True)

tts.tts_to_file(text=text, speaker_wav=voice_path, language="en", file_path=output_file)
```

```python
# ✅ Listen or Download the result
from IPython.display import Audio
Audio(output_file)
```
